# experiment 1: random perturbations

```python
with open(f'../results/{dataname}_{cumulative}_exp1_{p}.lls', 'rb') as infile:
    loaded = pickle.load(infile)
```

```python
xs = np.asarray(sorted(loaded.keys()))
ys = [loaded[0] for _ in xs]
yavg = [np.mean(loaded[x]) for x in xs]
ymax = [np.max(loaded[x]) for x in xs]
ymin = [np.min(loaded[x]) for x in xs]
```

```python
with plt.style.context(['ipynb', 'use_mathtext', 'colors5-light']):
    plt.title(f'Experiment 1: {dataname}, lookback {cumulative}')
    plt.xlabel('percentage (%) of total edges randomly rewired')
    plt.ylabel('log likelihood')
    plt.plot(xs * 100, ys)
    plt.plot(xs * 100, yavg)
    plt.fill_between(xs * 100, ymax, ymin, alpha=0.2)
    plt.legend(['true LL', 'mean of perturbed'])
    plt.ticklabel_format(style='plain')
    plt.savefig(f'../figures/exp1_{dataname}_{cumulative}.svg')
```

```python
rho = stats.spearmanr(xs, yavg)
print(rho)
print(rho.correlation)
```

# experiment 3: sequential perturbations

```python
dataname = 'email-dnc'
with open(f'../results/{dataname}{cumulative}_exp3.lls', 'rb') as infile:
    loaded = pickle.load(infile)
years = list(range(len(loaded)))
```

```python
xs = np.asarray(sorted(loaded[1].keys()))
ys = [loaded[idx] for idx in years[1:]]
yavgs = [[np.mean(ys[idx][x]) for x in xs] for idx in years[:-1]]
ymins = [[np.min(ys[idx][x]) for x in xs] for idx in years[:-1]]
ymaxs = [[np.max(ys[idx][x]) for x in xs] for idx in years[:-1]]
```

```python
with plt.style.context(['ipynb', 'use_mathtext', 'colors5-light']):
    plt.title(f'Experiment 3: {dataname}, lookback {cumulative}')
    plt.xlabel('percentage (%) of total edges randomly rewired')
    plt.ylabel('log likelihood')
    for year, triple in enumerate(zip(yavgs, ymins, ymaxs)):
        yavg, ymin, ymax = triple
        plt.plot(xs * 100, yavg, label=f't={year+1}')
        plt.fill_between(xs * 100, ymax, ymin, alpha=0.2)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.ticklabel_format(style='plain')
    plt.savefig(f'../figures/exp3_{dataname}_{cumulative}.svg')
```

# 3D Plots: experiment 1

```python
dataname = 'fb-messages'
p = 20

#lookbacks = [0, 1, 2, 3, 4, 5, 6, 10]
lookbacks = list(range(11))
args = ['ipynb',
        'use_mathtext',
        'colors5-light',
       ] if False else []

rhos = dict()

with plt.style.context(args):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.ticklabel_format(style='plain')

    for lookback in lookbacks:
        with open(f'../results/{dataname}_{lookback}_exp1_{p}.lls', 'rb') as infile:
            loaded = pickle.load(infile)

            xs = np.asarray(sorted(loaded.keys()))
            ys = [loaded[0] for _ in xs]
            yavg = [np.mean(loaded[x]) for x in xs]
            ymax = [np.max(loaded[x]) for x in xs]
            ymin = [np.min(loaded[x]) for x in xs]
            
            rho = stats.spearmanr(xs, yavg)
            rhos[lookback] = round(rho.correlation, 4)

            ax.plot([lookback for _ in xs], xs, yavg)

    ax.set_title(f'Experiment 1: {dataname}')
    ax.set_xlabel('amount of lookback')
    ax.set_ylabel('% of edges rewired')
    ax.set_zlabel('log likelihood')
    ax.legend([f'lookback = {x},    rho = {rhos[x]}' for x in lookbacks], bbox_to_anchor=(1.2, 1.05))
    plt.gca().invert_xaxis()
    plt.savefig(f'../figures/exp1_{dataname}_3D.svg', bbox_inches='tight')
    
```
